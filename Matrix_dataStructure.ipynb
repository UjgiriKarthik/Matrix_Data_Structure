```python
class matrix:
    def __init__(self):
        self._matrix = []
        self.row = int(input("Enter the order of square matrix: "))
        for i in range(self.row):
            a = []
            for j in range(self.row):
                a.append(int(input(f"Enter element at row {i+1}, column {j+1}: ")))
            self._matrix.append(a)
        print("Matrix created:")
        for i in self._matrix:
            print(i)

    def __str__(self):
        return "\n".join(" ".join(map(str, row)) for row in self._matrix)

    def __add__(self, other):
        result = []
        for i in range(self.row):
            row = []
            for j in range(self.row):
                row.append(self._matrix[i][j] + other._matrix[i][j])
            result.append(row)
        print("Addition Result:")
        for i in result:
            print(i)

    def __sub__(self, other):
        result = []
        for i in range(self.row):
            row = []
            for j in range(self.row):
                row.append(self._matrix[i][j] - other._matrix[i][j])
            result.append(row)
        print("Subtraction Result:")
        for i in result:
            print(i)

    def __mul__(self, other):
        result = [[0]*self.row for _ in range(self.row)]
        for i in range(self.row):
            for j in range(self.row):
                for k in range(self.row):
                    result[i][j] += self._matrix[i][k] * other._matrix[k][j]
        print("Multiplication Result:")
        for i in result:
            print(i)

    def __eq__(self, other):
        if self._matrix == other._matrix:
            print("Both matrices are Equal")
            return True
        else:
            print("Both matrices are NOT Equal")
            return False

    def determinant_of_matrix(self):
        if self.row == 2:
            return self._matrix[0][0]*self._matrix[1][1] - self._matrix[0][1]*self._matrix[1][0]

        if self.row == 3:
            return (
                self._matrix[0][0]*(self._matrix[1][1]*self._matrix[2][2] - self._matrix[1][2]*self._matrix[2][1]) -
                self._matrix[0][1]*(self._matrix[1][0]*self._matrix[2][2] - self._matrix[1][2]*self._matrix[2][0]) +
                self._matrix[0][2]*(self._matrix[1][0]*self._matrix[2][1] - self._matrix[1][1]*self._matrix[2][0])
            )

        print("Determinant supported only for 2x2 and 3x3 matrices")
        return None

    def transpose_of_matrix(self):
        transpose = []
        for i in range(self.row):
            row = []
            for j in range(self.row):
                row.append(self._matrix[j][i])
            transpose.append(row)

        print("Transpose:")
        for i in transpose:
            print(i)

    def scalar_multiplication_of_matrix(self):
        scalar = int(input("Enter scalar value: "))
        print("Scalar Multiplication Result:")
        for i in range(self.row):
            for j in range(self.row):
                self._matrix[i][j] *= scalar
        for i in self._matrix:
            print(i)


print("\nCreate First Matrix")
m1 = matrix()

print("\nCreate Second Matrix")
m2 = matrix()

while True:
    print("1. Addition")
    print("2. Subtraction")
    print("3. Multiplication")
    print("4. Equality Check")
    print("5. Determinant of Matrix 1")
    print("6. Transpose of Matrix 1")
    print("7. Scalar Multiplication of Matrix 1")
    print("8. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        m1 + m2
    elif choice == 2:
        m1 - m2
    elif choice == 3:
        m1 * m2
    elif choice == 4:
        m1 == m2
    elif choice == 5:
        print("Determinant:", m1.determinant_of_matrix())
    elif choice == 6:
        m1.transpose_of_matrix()
    elif choice == 7:
        m1.scalar_multiplication_of_matrix()
    elif choice == 8:
        print("Program Ended")
        break
    else:
        print("Invalid choice")
```
